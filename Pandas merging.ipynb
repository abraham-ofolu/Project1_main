```python
person_info = {
    "name": ["John", "sharoz", "Jane", "Jeff", "Tom"],
    "email": ["j@gmail.com", "s@gmail.com","Ja@gmail.com","je@gmail.com","T@gmail.com"],
    "age": [21,20,21,20,19]
}
order_info = {
    "customer name":["John", "sharoz", "Jane", "Grace"],
    "item":["Chips", "Laptop", "TV", "Cooler"],
    "cost":[50, 65, 55, 100]
}
df1 = pd.DataFrame(person_info)
df2 = pd.DataFrame(order_info)
```

```python
df2
```

```python
pd.merge(df1,df2, left_on="name", right_on="customer name", how="inner")
```
